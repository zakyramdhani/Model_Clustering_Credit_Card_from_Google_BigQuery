# Import Library 

```python
import pickle
import pandas as pd
import numpy as np
```

# Load Model

```python
# meload model
with open('model_scaler_cluster.pkl', 'rb') as file_1:
  model_scaler = pickle.load(file_1)

with open('model_kmeans.pkl', 'rb') as file_2:
  model_kmeans = pickle.load(file_2)

with open('model_pca.pkl', 'rb') as file_3:
  model_pca = pickle.load(file_3)
```

# New Data Random

```python
# membuat kolom dengan value acak hingga 100 row
np.random.seed(0)
BALANCE = np.random.uniform(0,18495,100)
BALANCE_FREQUENCY = np.random.uniform(0,1,100)
PURCHASES = np.random.uniform(0,41050,100)
ONEOFF_PURCHASES = np.random.uniform(0,40624,100)
INSTALLMENTS_PURCHASES = np.random.uniform(0,14686,100)
CASH_ADVANCE = np.random.uniform(0,47137,100)
PURCHASES_FREQUENCY = np.random.uniform(0,1,100)
ONEOFF_PURCHASES_FREQUENCY = np.random.uniform(0,1,100)
PURCHASES_INSTALLMENTS_FREQUENCY = np.random.uniform(0,1,100)
CASH_ADVANCE_FREQUENCY = np.random.uniform(0,1.5,100)
CASH_ADVANCE_TRX = np.random.uniform(0,123,100)
PURCHASES_TRX = np.random.uniform(0,347,100)
CREDIT_LIMIT= np.random.uniform(50,30000,100)
PAYMENTS = np.random.uniform(0.049513,50721,100)
MINIMUM_PAYMENTS = np.random.uniform(0.019163,76406,100)
PRC_FULL_PAYMENT = np.random.uniform(0,1,100)
TENURE = np.random.choice(['6','7','8','9','10','11','12'],100)
```

```python
# membuat nama kolom dalam bentuk dataFrame dengan data yang sudah dibuat secara acak
Inference_df = pd.DataFrame({
    'bal' : BALANCE,
    'bf': BALANCE_FREQUENCY,
    'pur': PURCHASES,
    'op': ONEOFF_PURCHASES,
    'ip' : INSTALLMENTS_PURCHASES,
    'ca': CASH_ADVANCE,
    'pf': PURCHASES_FREQUENCY,
    'opf': ONEOFF_PURCHASES_FREQUENCY,
    'pif' : PURCHASES_INSTALLMENTS_FREQUENCY,
    'caf': CASH_ADVANCE_FREQUENCY,
    'cat': CASH_ADVANCE_TRX,
    'pt': PURCHASES_TRX,
    'cal': CREDIT_LIMIT,
    'pay': PAYMENTS,
    'mp': MINIMUM_PAYMENTS,
    'pfp': PRC_FULL_PAYMENT,
    'ten': TENURE,

})

Inference_df
```

### Scaling New Data Random with Model

```python
# scaling model yang sudah dibuat lalu di fit dengan data baru
Inference_scaler = model_scaler.fit_transform(Inference_df)
Inference_scaler
```

### Dimentional Reduction New Data Random with model

```python
# reduksi dimensi dari model yang sudah di scale
Inference_pca = model_pca.fit_transform(Inference_scaler)
Inference_pca
```

### Predict New Data Random with model KMeans

```python
# memprediksi data baru dengan model yang sudah dibuat
Inference_cluster = model_kmeans.predict(Inference_pca)
Inference_cluster
```

### Predict Result

```python
# membuat data frame hasil prediksi
Inference_cluster_df = pd.DataFrame(Inference_cluster, columns=['Clustering'])
Inference_cluster_df
```

```python
# menggabungkan data baru dan hasil prediksi clustering
Inference_table = pd.concat([Inference_df, Inference_cluster_df], axis=1)
Inference_table
```
